####  Using Matplotlib, plot lineplots that display crypto daily prices on one single figure

```python
prices_df.plot(x = 'Date', y = ['BTC-USD Price', 'ETH-USD Price', 'LTC-USD Price'], label = ['BTC-USD Price', 'ETH-USD Price', 'LTC-USD Price'], figsize = (15,10), linewidth = 2)
plt.ylabel('Price [$]')
plt.xlabel('Date')
plt.title('Crypto daily prices')
plt.grid()
```

#### Using Matplotlib, plot crypto daily prices on multiple subplots

```python
plt.figure(figsize = (20,10))

plt.subplot(3,1,1)  # 3 row, 1 column, 1. plot
plt.plot(prices_df['BTC-USD Price'], 'r--')
plt.grid()

plt.subplot(3,1,2)  # 3 row, 1 column, 2. plot
plt.plot(prices_df['ETH-USD Price'], 'b--')
plt.grid()

plt.subplot(3,1,3)  # 3 row, 1 column, 3. plot  
plt.plot(prices_df['LTC-USD Price'], 'w--')
plt.grid()
```

#### Assume that you became bullish on XRP and decided to allocate 60% of your assets in it. You also decided to equally divide the rest of your assets in other coins (BTC, LTC, ADA, and ETH). USing Matplotlib, plot a pie chart that shows these allocations. Use explode" attribute to increase the separation between BTC and the rest of the portfolio.

```python
#order matters, it has to be connected to labels, 20=g=0=AAPL, etc.
values  = [60, 10, 10, 10, 10]  
colors  = ['g', 'r', 'y', 'b', 'm']
explode = [0, 0.2, 0, 0, 0]
labels  = ['XRP', 'BTC', 'LTC', 'ADA', 'ETH']

plt.figure(figsize = (6,6))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('Crypto Portfolio Pie Chart')
```

#### Using Matplotlib, plot the histogram for ETH returns using 20 bins with red color

```python
returns_df
```

```python
mu = returns_df['ETH'].mean() #mean of distribution
sigma = returns_df['ETH'].std() #standard deviation of distribution

num_bins = 40

plt.figure(figsize = (20,15))

returns_df['ETH'].plot.hist(bins = 20, alpha = 0.5, facecolor = 'r', figsize = (12,6))
        
plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu) + ', sigma = ' +str(sigma));
```

#### Using Seaborn, plot a heatmap that shows the correlations between crypto daily returns.

```python
plt.figure (figsize = (8, 8))
sns.heatmap(returns_df.corr(), annot = True)
```
